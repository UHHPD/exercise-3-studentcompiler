### Exercise 1: Maximum likelihood method

b) plotting likelihood.txt

```python
import numpy as np
import matplotlib.pyplot as plt


data_L = np.loadtxt('likelihood.txt')


x = data_L[:, 0]
y = data_L[:, 1]

plt.xlabel("μ")
plt.ylabel("L(μ)")
plt.title("Maximum likelihood L as a function of μ")

plt.plot(x, y,'r--')
plt.show()

print("The maximum value for L(μ) is when μ =",x[np.argmax(y)])
```

The maximum is at at μ ≈ 3.1 which corresponds to the real mean µ = 3.11538

```python
data_delnll = np.loadtxt('deltanll.txt')

# mu
x = data_delnll[:, 0]
# 2L(mu) - 2L(3.11538)
y = data_delnll[:, 1]

N = 234
var = 2.74038

plt.xlabel("μ")
plt.title("Plotting 2L(μ)-2L(3.11538)")
plt.plot(x, y,'r--')
plt.show()

L_minimum = np.min(y)
best_estimator_μ = x[np.argmin(y)]
print("The minimum value is when μ̂ =", best_estimator_μ)

uncertainty_range = x[y < 1.0]
print(f"The uncertainty interval for μ: {uncertainty_range[0]} < μ < {uncertainty_range[-1]}")

sample_mean = np.sqrt(var/N)

print("uncertainty on the sample mean ",sample_mean)
```
